```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generating sample data
np.random.seed(0)
num_students = 1000

# Random number of lectures attended (from 0 to 100)
lectures_attended = np.random.randint(0, 101, num_students)

# Random hours spent on final project (from 0 to 50)
hours_spent_on_project = np.random.randint(0, 51, num_students)

# Generating pass/fail labels based on a simple rule
pass_fail_labels = (lectures_attended > 50) & (hours_spent_on_project > 25)

# Creating feature matrix X and labels y
X = np.column_stack((lectures_attended, hours_spent_on_project))
y = pass_fail_labels.astype(int)

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Feature scaling (standardization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the neural network model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.1)

# Evaluate the model on test data
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Test Accuracy: {accuracy}")
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Generating sample data
np.random.seed(0)
num_students = 1000

# Random number of lectures attended (from 0 to 100)
lectures_attended = np.random.randint(0, 101, num_students)

# Random hours spent on final project (from 0 to 50)
hours_spent_on_project = np.random.randint(0, 51, num_students)

# Random number of hours studied (from 0 to 150)
hours_studied = np.random.randint(0, 151, num_students)

# Random number of extracurricular activities (from 0 to 5)
extracurricular_activities = np.random.randint(0, 6, num_students)

# Generating academic scores based on a formula
academic_score = (0.3 * lectures_attended +
                  0.2 * hours_spent_on_project +
                  0.4 * hours_studied +
                  0.1 * extracurricular_activities +
                  np.random.normal(0, 5, num_students))

# Creating feature matrix X and target y
X = np.column_stack((lectures_attended, hours_spent_on_project, hours_studied, extracurricular_activities))
y = academic_score

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Feature scaling (standardization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Define the neural network model for regression
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)  # No activation function for regression
])

# Compile the model
model.compile(optimizer='adam',
              loss='mean_squared_error',  # Use mean squared error for regression
              metrics=['mean_absolute_error'])  # Additional metric: mean absolute error

# Train the model
history = model.fit(X_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.1)

# Evaluate the model on test data
loss, mae = model.evaluate(X_test_scaled, y_test)
print(f"Test Mean Absolute Error: {mae}")
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize the data
X_train, X_test = X_train / 255.0, X_test / 255.0

# Define the model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer=Adam(),
              loss=SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test accuracy: {accuracy}')
```

```python
import numpy as np

class DeepNeuralNetwork:
    def __init__(self, layer_dims, learning_rate=0.01, n_iters=1000):
        self.layer_dims = layer_dims
        self.lr = learning_rate
        self.n_iters = n_iters
        self.params = self.initialize_parameters()

    def initialize_parameters(self):
        np.random.seed(1)
        params = {}
        L = len(self.layer_dims)
        for l in range(1, L):
            params[f'W{l}'] = np.random.randn(self.layer_dims[l], self.layer_dims[l-1]) * 0.01
            params[f'b{l}'] = np.zeros((self.layer_dims[l], 1))
        return params

    def sigmoid(self, Z):
        return 1 / (1 + np.exp(-Z))

    def sigmoid_derivative(self, Z):
        return self.sigmoid(Z) * (1 - self.sigmoid(Z))

    def forward(self, X):
        caches = {}
        A = X.T
        L = len(self.params) // 2
        for l in range(1, L + 1):
            A_prev = A
            W = self.params[f'W{l}']
            b = self.params[f'b{l}']
            Z = np.dot(W, A_prev) + b
            A = self.sigmoid(Z)
            caches[f'Z{l}'] = Z
            caches[f'A{l-1}'] = A_prev
        caches[f'A{L}'] = A
        return A, caches

    def backward(self, X, Y, caches):
        grads = {}
        L = len(self.params) // 2
        m = X.shape[0]
        Y = Y.reshape(-1, 1).T

        A_L = caches[f'A{L}']
        dA_L = - (np.divide(Y, A_L) - np.divide(1 - Y, 1 - A_L))

        for l in reversed(range(1, L + 1)):
            dZ = dA_L * self.sigmoid_derivative(caches[f'Z{l}'])
            A_prev = caches[f'A{l-1}']
            W = self.params[f'W{l}']
            grads[f'dW{l}'] = 1/m * np.dot(dZ, A_prev.T)
            grads[f'db{l}'] = 1/m * np.sum(dZ, axis=1, keepdims=True)
            dA_L = np.dot(W.T, dZ)

        return grads

    def update_parameters(self, grads):
        L = len(self.params) // 2
        for l in range(1, L + 1):
            self.params[f'W{l}'] -= self.lr * grads[f'dW{l}']
            self.params[f'b{l}'] -= self.lr * grads[f'db{l}']

    def fit(self, X, Y):
        for i in range(self.n_iters):
            A_L, caches = self.forward(X)
            grads = self.backward(X, Y, caches)
            self.update_parameters(grads)

    def predict(self, X):
        A_L, _ = self.forward(X)
        predictions = (A_L > 0.5).astype(int)
        return predictions.T

# Usage
if __name__ == "__main__":
    # Create some dummy data
    np.random.seed(42)
    X = np.random.randn(1000, 10)
    y = (np.sum(X, axis=1) > 0).astype(int)

    # Define the network architecture
    layer_dims = [10, 5, 3, 1]  # 3-layer network

    # Initialize and train the network
    dnn = DeepNeuralNetwork(layer_dims=layer_dims, learning_rate=0.01, n_iters=1000)
    dnn.fit(X, y)

    # Make predictions
    predictions = dnn.predict(X)
    accuracy = np.mean(predictions == y.reshape(-1, 1))
    print(f"Deep Neural Network accuracy: {accuracy:.2f}")
```
